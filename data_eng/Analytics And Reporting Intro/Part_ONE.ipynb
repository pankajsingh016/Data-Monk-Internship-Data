```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# 1. Histogram of Tip Amounts
plt.figure(figsize=(7,5))
sns.histplot(df["tip"], bins=20, kde=True)
plt.title("Distribution of Tip Amounts")
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Compare Total Bills by Time of Day
plt.figure(figsize=(7,5))
sns.barplot(x="time", y="total_bill", data=df, estimator=sum, ci=None)
plt.title("Total Bills: Lunch vs Dinner")
plt.xlabel("Time of Day")
plt.ylabel("Total Bill (Sum)")
plt.show()
```

```python
# 3. Box Plot of Total Bill by Day
plt.figure(figsize=(7,5))
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Total Bill Distribution by Day")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
# 4. Heatmap (using flights dataset)
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(10,7))
sns.heatmap(flights, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Flights Heatmap (Passengers)")
plt.show()
```
